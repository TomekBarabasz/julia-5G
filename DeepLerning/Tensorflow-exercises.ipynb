```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
```

```python
celcius    = np.array([-40, -10,  0,  8, 15, 22,  38], dtype=float)
fahrenheit = np.array([-10,  14, 32, 46, 59, 72, 100], dtype=float)
```

```python
model = keras.Sequential([
    layers.Dense(units=1, input_shape=[1]) #activation='relu'
])
```

```python
#model.compile(optimizer='adam', loss='mse')
model.compile(loss='mse', optimizer=keras.optimizers.Adam(0.1))
```

```python
history = model.fit(celcius,fahrenheit, epochs=800, verbose=False)
```

```python
plt.plot(history.history['loss'])
plt.xlabel('epoch number')
plt.ylabel('loss')
plt.show()
```

```python
c = np.arange(-40,40,0.25)
f = c * 1.8 + 32
fp = model.predict(c)
```

```python
d=f.T-fp.T
d.shape
```

```python
plt.plot(c, d.T)
plt.xlabel('celcius')
plt.ylabel('error')
plt.show()
```
